#1. Extracting Tesla Stock Data Using yfinance

```python
import yfinance as yf
import pandas as pd

# 1. Create the ticker object and pull history
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# 2. Reset the index and show the first five rows
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
tesla_data.tail()
```

#2. Extracting Tesla Revenue Data Using Webscraping

```python
import yfinance as yf
import pandas as pd

# 1) Pull the ticker
gme = yf.Ticker("TSLA")

# 2) Grab quarterly Income Statement
q_fin = tesla.quarterly_financials

# 3) Extract just “Total Revenue” and transpose so each row is a quarter
gme_revenue = (
    q_fin.loc["Total Revenue"]
         .to_frame(name="Revenue")
         .reset_index()
         .rename(columns={"index":"Date"})
)

# 4) Convert to millions (if you want)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"] / 1e6

# 5) Preview
gme_revenue.tail()
```

```python
gme = yf.Ticker("GME")
gme_q_fin = gme.quarterly_financials

gme_revenue = (
    gme_q_fin.loc["Total Revenue"]
             .to_frame(name="Revenue")
             .reset_index()
             .rename(columns={"index":"Date"})
)
gme_revenue["Revenue"] = gme_revenue["Revenue"] / 1e6
gme_revenue.tail()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def make_graph(stock_df, revenue_df, title):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                        subplot_titles=("Stock Price", "Revenue"))

    fig.add_trace(
        go.Scatter(x=stock_df['Date'], y=stock_df['Close'], name="Close Price"),
        row=1, col=1
    )

    fig.add_trace(
        go.Bar(x=revenue_df['Date'], y=revenue_df['Revenue'], name="Revenue"),
        row=2, col=1
    )

    fig.update_layout(height=600, title_text=title)
    return fig

# 5. Plot Tesla
fig_tsla = make_graph(tesla_data, tesla_revenue, "Tesla: Stock Price vs Revenue")
fig_tsla.show()
```

```python
# 6. Plot GameStop
fig_gme = make_graph(gme_data, gme_revenue, "GameStop: Stock Price vs Revenue")
fig_gme.show()
```
